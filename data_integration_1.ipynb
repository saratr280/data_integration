```python
census_data = pd.read_csv("acs2017_census_tract_data.csv")
census_data.head(10)
```

```python
covid_data = pd.read_csv("COVID_county_data.csv")
covid_data.head(10)
```

```python
covid_data = covid_data.rename(columns={"state": "State", "county": "County"})
covid_data.head()
```

```python
census_data.groupby(["State","County"]).sum()
```

```python
covid_data.groupby(["State","County"]).sum()
```

```python
joined_data = pd.merge(census_data.groupby(["State","County"]).sum(),covid_data.groupby(["State","County"]).sum(),how="inner",on=["State","County"])
joined_data
```
